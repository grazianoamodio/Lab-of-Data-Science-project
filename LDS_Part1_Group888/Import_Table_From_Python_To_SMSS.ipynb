```python
# IMPORT DI GUN TABLE
input_csv_file = 'C:\\Users\\marco\\gun_table.csv'


# Definizione della tabella "gun" nel database
gun_table_definition = '''
CREATE TABLE guns (
    gun_id INT PRIMARY KEY,
    gun_stolen VARCHAR(255),
    gun_type VARCHAR(255)
);
'''

# Esecuzione della query per creare la tabella
cursor.execute(gun_table_definition)
conn.commit()

# Inserimento dei dati nella tabella "gun"
with open(input_csv_file, 'r') as gun_csv:
    gun_reader = csv.DictReader(gun_csv)
    for row in gun_reader:
        gun_id = row['gun_id']
        gun_stolen = row['gun_stolen']
        gun_type = row['gun_type']

        # Query di inserimento dei dati
        insert_query = f"INSERT INTO guns (gun_id, gun_stolen, gun_type) VALUES ({gun_id}, '{gun_stolen}', '{gun_type}');"
        cursor.execute(insert_query)

# Committa le modifiche al database
conn.commit()



print("Importazione dei dati nella tabella 'gun' completata.")
```

```python
# IMPORT DI GEO TABLE
input_csv_file = 'C:\\Users\\marco\\geo_table_giusta.csv'


# Definizione della tabella "geo" nel database
geo_table_definition = '''
CREATE TABLE geography (
    geo_id INT PRIMARY KEY,
    latitude VARCHAR(255),
    longitude VARCHAR(255),
    city VARCHAR(255),
    state_id VARCHAR(255),
    state_name VARCHAR(255)

);
'''

# Esecuzione della query per creare la tabella
cursor.execute(geo_table_definition)
conn.commit()

# Inserimento dei dati nella tabella "geo"
with open(input_csv_file, 'r') as geo_csv:
    geo_reader = csv.DictReader(geo_csv)
    for row in geo_reader:
        geo_id = row['geo_id']
        latitude = row['latitude']
        longitude = row['longitude']
        city = row['city']
        state_id = row['state_id']
        state_name = row['state_name']

        # Query di inserimento dei dati con parametri
        insert_query = "INSERT INTO geography (geo_id, latitude, longitude, city, state_id, state_name) VALUES (?, ?, ?, ?, ?, ?);"
        cursor.execute(insert_query, geo_id, latitude, longitude, city, state_id, state_name)

# Committa le modifiche al database
conn.commit()




print("Importazione dei dati nella tabella 'geo' completata.")
```
